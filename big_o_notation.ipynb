```python
import matplotlib.pyplot as plt
import math
import pandas as pd
import numpy as np
```

```python
# 5. Make a new Jupyter notebook named big_o_notation.ipynb
#       Title your chart "Big O Notation"
#       Label your x axis "Elements"
#       Label your y axis "Operations"
#       Label your curves or make a legend for the curves
#       Use LaTex notation where possible
#    Curves to graph
#       y = 0n + 1 and label the curve "O(1)"
#       y = log(n) and label the curve "O(log n)"
#       y = n and label the curve "O(n)"
#       y = n * log(n) and label it "O(n log n)"
#       y = n^2 and label it "O(n^2)"
#       y = 2^n and label it "O(2^n)"
#       y = n! and label it "O(n!)"
#       y = n^n and label it "O(n^n)"
```

```python
plt.figure(figsize = (15,5))

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

x = list(range(1,15))

y1 = [0*n + 1 for n in x]
plt.plot(x,y1, c='tomato', label = "O(1)")

y2 = [math.log(n) for n in x]
plt.plot(x,y2, c='black', label = "$O\log(n)$")

y3 = [n for n in x]
plt.plot(x,y3, c ='crimson', label = "O(n)")

y4 = [n * math.log(n) for n in x] 
plt.plot(x,y4, c ='goldenrod', label = "$O(n\log(n))$")

y5 = [n**2 for n in x]
plt.plot(x,y5, c ='blue', label = "$O(n^2)$")

y6 = [2**n for n in x]
plt.plot(x,y6, c = 'limegreen', label = "$O(2^n)$")

y7 = [math.factorial(n) for n in x]
plt.plot(x,y7, c = 'darkorange', label = "$On!$")

y8 = [n**n for n in x]
plt.plot(x,y8, c = 'pink', label = "$O(n^n)$")

plt.ylim(0,25)

plt.legend()
plt.show()
```

```python
# y = 0n + 1
x = pd.Series(range(1, 14))
y = 0*x + 1
```

```python
# y = log(n)
y1 = np.log(x)
```

```python
# y = n
y2 = x
```

```python
# y = n * log(n)
y3 = x * np.log(x)
```

```python
# n squared
y4 = x*x
```

```python
# 2^n
y5 = 2**x
```

```python
# n!
# np.math.factorial
# pandas_series.apply(function_body)
y6 = x.apply(np.math.factorial)
```

```python
# n^n
y7 = x**x
```

```python
plt.title("Big O Notation")
plt.plot(x, y, label="0(n) + 1")
plt.plot(x, y1, label="log(n)")
plt.plot(x, y2, label="y = n")
plt.plot(x, y3, label="n * log(n)")
plt.plot(x, y4, label="x**2")
plt.plot(x, y5, label="2**x")
plt.plot(x, y6, label="factorial(x)")
# plt.plot(x, y7, label="x^x")
plt.legend(loc='upper right')
plt.xlim(0,20)
plt.ylim(0,20)
plt.show()
```
