```python
import matplotlib.pyplot as plt
import math
import random
import pandas as pd
import numpy as np
```

```python
# 1. Use matplotlib to plot the following equation: y = x^2 − x + 2
#    You'll need to write the code that generates the x and y points.
#    Add an anotation for the point 0, 0, the origin.

x = list(range(-15,15))
y = [n**2 - n + 2 for n in x]

plt.plot(x,y)
plt.text(-1,8, '(0,0)', fontsize=10)
plt.title('$y=x^2 - x +2$')
plt.xlabel('x')
plt.ylabel('y')
plt.annotate('the origin', xy = (0,20), xytext = (0,125), arrowprops={})

plt.show()

```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
#       y = √x    
#       y = x^3 
#       y = 2^x
#       y = 1/(x + 1)
# You can use functions from the math module to help implement some of the equations above.
```

```python
x = list(range(0,10))
y = [(n**(1/2)) for n in x]

plt.plot(x,y, c='green')
plt.title('$y=\sqrt{x}$')

plt.show()
```

```python
x = list(range(-5,5))
y = [n**3 for n in x]

plt.plot(x, y, c='navy')
plt.title("$y=x^3$")

plt.show()
```

```python
x = list(range(-5,10))
y = [2**n for n in x]

plt.plot(x,y, c='slategray')
plt.title("$y=2^x$")

plt.show()
```

```python
x = list(range(-12,-1))
y = [1/(n+1) for n in x] 

plt.plot(x,y, c='lightblue')
plt.title("y=1/(x+1)")

plt.show()
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.

plt.figure(figsize = (10,10))
plt.subplot (2,2,1)
x = list(range(0,10))
y = [(n**(1/2)) for n in x]
plt.plot(x,y, c='green')
plt.title('$y=\sqrt{x}$')


plt.subplot(2,2,2)
x2 = list(range(-5,5))
y2 = [n**3 for n in x2]
plt.plot(x2, y2, c='navy')
plt.title("$y=x^3$")


plt.subplot(2,2,3)
x3 = list(range(-5,6))
y3 = [2**n for n in x3]
plt.plot(x3,y3, c='slategray')
plt.title("$y=2^x$")


plt.subplot(2,2,4)
x4 = list(range(-12,-1))
y4 = [1/(n+1) for n in x4] 
plt.plot(x4,y4, c='lightblue')
plt.title("y=1/(x+1)")


plt.tight_layout()

plt.show()
```

```python
# 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a
#    different color for the points. Be sure to include a legend and an appropriate title for the figure.
```

```python
plt.figure(figsize = (8,8))

x = list(range(0,10))
y = [(n**(1/2)) for n in x]
plt.plot(x, y, c='green', label = '$y=\sqrt{x}$')


x2 = list(range(-5,5))
y2 = [n**3 for n in x2]
plt.plot(x2, y2, c='navy', label = '$y = x^3$')


x3 = list(range(-5,6))
y3 = [2**n for n in x3]
plt.plot(x3, y3, c='slategray', label = '$y= 2^x$')


x4 = list(range(-12,-1))
y4 = [1/(n+1) for n in x4] 
plt.plot(x4, y4, c = 'lightblue', label = 'y = 1/(x+1)')


plt.ylim(-15,20)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Combined Graphs")

plt.legend()

plt.show()

```
