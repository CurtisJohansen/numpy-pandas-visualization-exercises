```python
# 4. Using the chipotle dataset from the previous exercise

from env import host, user, password

def get_db_url(database):
    url = f'mysql+pymysql://{user}:{password}@{host}/{database}'
    return url

chipotle_df = pd.read_sql('SELECT * FROM orders', get_db_url('chipotle'))
chipotle_df
```

```python
chipotle_df.dtypes
```

```python
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
```

```python
# make the prices a float and remove $

chipotle_df.item_price = chipotle_df.item_price.str.replace(r'\$','',regex=True).astype(float)
```

```python
# find 4 most popular items

popular_items = chipotle_df.groupby('item_name').item_price.sum().sort_values(ascending = False).head(4)
popular_items
```

```python
# revenue from the 4 most popular items

revenue = chipotle_df[chipotle_df.item_name.isin(popular_items.index)].groupby('item_name').item_price.sum()
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
# create a bar chart that shows the 4 most popular items and the revenue produced by each.

sns.barplot(x = 'item_name', y = 'item_price', data = revenue)
plt.title('The 4 most popular items and revenue')
```
